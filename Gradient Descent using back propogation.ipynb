```python
#Solution2.1
import numpy as np

def mse(y,y0):
    if len(y) == len (y0):
        se=(y0-y)**2;
    mse=np.sum(se)/len(y0);
    return(mse);  

a=np.array([1,2,3])
b=np.array([4,5,8])
mserr=mse(b,a)
print("MSE Values : %s" % mserr)
```

```python
#Solution 2.4
w= 0.5;
b= 0.5
num_example=x.shape[0]
#Initial value for the Gradient Descent Parameter
step_size = 1e-3 #Also called learning rate
Loss3=np.zeros([200])
# gradient descent loop
for i in range(200):
    # evaluate class scores, [N x K]
    scores = np.dot(x, w) + b 
    # compute the class probabilities
    Loss3[i]= mse(y,scores)
    #print (Loss[i])    
    db=2*np.sum((scores-y))/num_example
    dw=2*np.sum((scores-y)*x)/num_example
    #if i % 10 == 0:
        #print ("iteration %d: loss %f" % (i, Loss[i]))
    # perform a parameter update
    w += -step_size * dw
    b += -step_size * db
    
print(w,b)
```
