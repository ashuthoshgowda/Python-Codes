```python
dummy=np.empty((1682,1682))
df_xij = pd.DataFrame(index=range(1,1683), columns=range(1,1683))
df_xij = df_xij.fillna(0)
df_xij.head(n=5)
```

```python
df1=df1.sort(['movie id','user_id'],ascending=[1,1])
```

```python
user_by_movie=([[]for _ in range(1682)])
```

```python
for i in range(df1.shape[0]):
    movie_id=df1.iat[i,0]
    user_id=df1.iat[i,1]
    rating=df1.iat[i,2]
    if(rating==1):
        user_by_movie[(movie_id-1)].append(user_id)
```

```python
len= df1.shape[0]


for j in range(1682):
    for k in range(1682): 
        a=np.array(user_by_movie[j])
        b=np.array(user_by_movie[k])
        l=0
        m=0
        count=0
                
        while(l<a.shape[0] and m<b.shape[0]):
            if(a[l]==b[m]):
                count+=1
                l+=1
                m+=1
            elif(a[l]>b[m]):
                m+=1
            else:
                l+=1
        df_xij.set_value(j+1,k+1,count)

```

```python
v=np.matrix(np.random.uniform(-0.7,0.7, size=(1682, 10)))
```

```python
v=np.matrix(np.random.uniform(-0.7,0.7, size=(1682, 10)))
reg=1e-3
learning_rate=1e-4
result=[]
iter=30
for i in range(30):
    vt=v.transpose()
    #print(at.shape)
    b=np.dot(v,vt)
    data_loss=b-np.array(df_xij)
    np.fill_diagonal(data_loss,0)
    dloss_square=np.square(data_loss)
    reg_loss=0.5*reg*np.sum(b*b)
    loss=np.sum(dloss_square)+reg_loss
    if i%3==0:
        print"iterartion %d: loss %f" %(i,loss)
        learning_rate*=0.9
    dv= np.dot(vt,data_loss)
    result.append(loss)
    dv+= reg*v.transpose()
    f= dv*learning_rate
    
    v-=f.transpose()
```

```python
plt.suptitle('Loss vs Iteration')
plt.xlabel('Loss')
plt.ylabel('No of Iterations')
plt.plot(range(iter),result);
plt.show();
```
